# Histogram of Numeric Features

```python
plt.figure(figsize=(6,4))
plt.hist(df['price'], bins=20)
plt.title("Histogram of Price")
plt.xlabel("Price")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.figure(figsize=(6,4))
plt.hist(df['mileage'], bins=20)
plt.title("Histogram of Mileage")
plt.xlabel("Mileage")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.figure(figsize=(6,4))
plt.hist(df['year'], bins=20)
plt.title("Histogram of Year")
plt.xlabel("Year")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.figure(figsize=(6,4))
plt.hist(df['engineSize'], bins=20)
plt.title("Histogram of Engine Size")
plt.xlabel("Engine Size")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.figure(figsize=(6,4))
plt.hist(df['mpg'], bins=20)
plt.title("Histogram of MPG")
plt.xlabel("MPG")
plt.ylabel("Frequency")
plt.show()
```

```python
# Findings:
# 1. Most cars are in the low to medium price range, while high-priced cars are fewer.
# 2. Mileage and engine size are mostly concentrated in the middle range.
# 3. Most vehicles are from recent years, indicating a newer car dataset.
# 4. MPG values are generally moderate to high, showing that many cars are fuel-efficient.
```

# Correlation Heatmap

```python
plt.figure(figsize=(8,6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()

# Observations:
# 1. Features with correlation close to +1 have a strong positive relationship.
# 2. Features with correlation close to -1 have a strong negative relationship.
# 3. Check which features are highly correlated with 'price'.
```
