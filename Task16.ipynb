```python
sns.histplot(df['age'])
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
sns.histplot(df["bmi"])
plt.title("BMI Distribution")
plt.xlabel("BMI")
plt.ylabel("Count")
plt.show()
```

```python
sns.histplot(df["children"])
plt.title("Children Distribution")
plt.xlabel("Children")
plt.ylabel("Count")
plt.show()
```

```python
sns.histplot(df["expenses"])
plt.title("expenses Distribution")
plt.xlabel("expenses")
plt.ylabel("Count")
plt.show()
```

```python
sns.countplot(x="sex", data=df)
plt.title("Sex Count")
plt.xlabel("Sex")
plt.ylabel("Count")
plt.show()
```

```python
sns.countplot(x="smoker", data=df)
plt.title("Smoker Count")
plt.xlabel("Smoker")
plt.ylabel("Count")
plt.show()
```

```python
sns.countplot(x="region", data=df)
plt.title("Region Count")
plt.xlabel("Region")
plt.ylabel("Count")
plt.show()
```

```python
sns.heatmap(df.corr(numeric_only=True))
plt.title("Heatmap")
plt.show()
```

```python
# Average expenses
print("Average expenses",df["expenses"].mean())

# Maximum & Minimum expenses
print("Maximum expenses:",df["expenses"].max())
print("Minimum expenses:",df["expenses"].min())

# Average expenses of smokers and non-smokers
print(df.groupby("smoker")["expenses"].mean())
```

```python
sns.boxplot(x="smoker", y="expenses", data=df)
plt.title("Expenses by Smoker")
plt.show()
```

```python
sns.boxplot(x="sex", y="expenses", data=df)
plt.title("Expenses by Sex")
plt.show()
```
